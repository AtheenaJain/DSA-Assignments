# **Reading Data**

```python
data=pd.read_excel('iris.xls')
```

# **Display Columns in the dataset**

```python
data.columns                          #SL<,SW,PL,PW,Classification are the columns in this dataset
```

# **Mean of the each columns**

```python
data.mean()          #mean of sepal width is high
```

# **Check null values**

```python
data.isna().sum()              #There is no null values
```

# **Visualisations**

# Distribution of petal length of Iris flower

```python
plt.hist(data['PL'])                        #Iris flowers of petal length 1 to 1.5 are more coomon,petal length of approximately 2.2 to 2.7 are not in this data
plt.show()
```

# Countplot of Classification

```python
sns.countplot(x=data.Classification)              #Data entries of all 3 classifications are same
```

# Pie Chart of Sepal Width mean based on Classification

```python
sw_data=data.groupby(by='Classification')['SW'].mean()
Class_data=data['Classification'].unique()
```

```python
plt.pie(sw_data,labels=Class_data,autopct='%1.1f%%')             #Sepal width mean of Iris-setosa is high
plt.show()
```
